```hy
(import os base64 [numpy :as np]
        [pandas :as pd]
        [matplotlib.pyplot :as plt]
        [matplotlib_inline [show]]
        )

(np.random.seed 0)

(plt.style.use "ggplot")
```

```hy
(setv x (np.linspace 0 1 10))
(setv y (+ (* 2 x) (* 0.1 (np.random.randn 10))))

(setv fig (plt.figure))
(setv ax (fig.add_subplot 111))
(ax.plot x y)

(show fig)
```

```hy
(setv fig (plt.figure))

(setv ax (fig.add_subplot 111))

(setv y (+ (* 2 x) 1))

(.plot ax x y)

(show fig)
```
